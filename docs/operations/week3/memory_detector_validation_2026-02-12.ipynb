# Memory + Detector Validation Notebook (Week 3)

Date: 2026-02-12

This notebook reproduces the key Week 3 validation checks:
- memory fidelity and retrieval trends vs wait time
- detector click and false-click behavior under stochastic settings
- deterministic-seed repeatability

```python
from photonstrust.physics.memory import simulate_memory
from photonstrust.physics.detector import simulate_detector

memory_cfg = {
    "t1_ms": 50.0,
    "t2_ms": 10.0,
    "store_efficiency": 0.95,
    "retrieval_efficiency": 0.8,
    "physics_backend": "analytic",
}
waits = [0.0, 1e5, 1e7, 1e9]
rows = [simulate_memory(memory_cfg, w) for w in waits]
for w, r in zip(waits, rows):
    print(w, r.fidelity, r.p_retrieve, r.diagnostics)
```

```python
arrivals = [float(v) for v in range(0, 4000, 10)]
det_cfg = {
    "pde": 0.55,
    "dark_counts_cps": 200.0,
    "jitter_ps_fwhm": 30.0,
    "dead_time_ns": 10.0,
    "afterpulsing_prob": 0.01,
    "afterpulse_delay_ns": 50.0,
    "time_bin_ps": 10.0,
    "seed": 1234,
}

a = simulate_detector(det_cfg, arrivals)
b = simulate_detector(det_cfg, arrivals)

print("p_click", a.p_click, "p_false", a.p_false)
print("deterministic", a.p_click == b.p_click, a.click_time_hist == b.click_time_hist)
```
