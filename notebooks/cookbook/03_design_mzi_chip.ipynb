# Designing a Mach-Zehnder Interferometer Chip

A **Mach-Zehnder Interferometer (MZI)** is a fundamental building block in photonic
integrated circuits. It splits light into two arms, applies a relative phase shift,
then recombines the beams. The output power depends on the phase difference, making
the MZI useful for switching, modulation, and sensing.

```
            ┌─────────┐       phase shift       ┌─────────┐
  in1 ──▶│ MMI 2x2 │──────φ────────▶│ MMI 2x2 │──▶ out1
  in2 ──▶│         │───────────────▶│         │──▶ out2
            └─────────┘                       └─────────┘
```

In this notebook we use PhotonsTrust's `design_pic` helper to simulate a simple MZI.

````python
from photonstrust.easy import design_pic
````

````python
result = design_pic(
    components=["mmi_2x2", "mmi_2x2"],
    connections=[{"from": [0, "out1"], "to": [1, "in1"]}],
)
print(result.summary())
````

## Understanding PIC Simulation Results

The summary shows the insertion loss, extinction ratio, and S-parameter matrix for
the designed circuit. You can inspect the full result dictionary for wavelength-dependent
data and component-level details.

````python
import json
details = result.as_dict()
print(json.dumps(details, indent=2, default=str))
````
