# Beyond Direct Links — Quantum Repeater Analysis

Direct point-to-point QKD is fundamentally limited by the **PLOB bound**: the secret-key
capacity of a lossy bosonic channel scales as $-\log_2(1-\eta)$ where $\eta$ is the
channel transmittance. For standard fibre (0.2 dB/km), this means key rates drop
exponentially and vanish around 300–400 km.

**Twin-field (TF) protocols** overcome this limit by having both parties send optical
fields to a central measurement node, effectively halving the channel loss exponent.
This notebook compares direct-link and repeater-compatible protocols to show where
twin-field variants excel.

```python
from photonstrust.easy import compare_protocols
```

```python
comp = compare_protocols(
    protocols=["bb84", "tf_qkd", "sns_tf_qkd", "pm_qkd"],
    distances={"start": 0, "stop": 300, "step": 15},
    include_uncertainty=False,
)
print(comp.summary())
```

```python
# Which protocol wins at each distance?
print("Distance  Winner")
print("-" * 30)
for d in [50, 100, 150, 200, 250, 300]:
    winner = comp.winner_at(d)
    print(f"  {d:4d} km: {winner}")
```

## The Twin-Field Advantage

At short distances (< 100 km), standard BB84 often provides the highest key rate due
to its simpler implementation and lower overhead. However, as distance increases
beyond 150–200 km, twin-field protocols like **TF-QKD** and **SNS-TF-QKD** surpass
BB84 because their key rate scales with the square root of the channel transmittance
rather than linearly.

**PM-QKD** (phase-matching) offers a simpler optical setup than full TF-QKD while
retaining much of the distance advantage.
