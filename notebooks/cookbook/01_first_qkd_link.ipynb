# Your First QKD Link Simulation

**Quantum Key Distribution (QKD)** allows two parties to generate a shared secret key
whose security is guaranteed by the laws of quantum mechanics. The most widely known
protocol is **BB84**, proposed by Bennett and Brassard in 1984. In BB84, single photons
are encoded in one of two conjugate bases; an eavesdropper is detected because
measuring a quantum state inevitably disturbs it, raising the observed error rate above
a safe threshold. In this notebook we simulate a complete BB84 fibre link using
PhotonsTrust's high-level API.

```python
from photonstrust.easy import simulate_qkd_link
```

```python
result = simulate_qkd_link(
    protocol="bb84",
    distance_km=50,
    band="c_1550",
    detector="snspd",
    include_uncertainty=False,
)
print(result.summary())
```

## Understanding the Results

The summary above shows the **secret-key rate** (bits/s), the **QBER** (quantum bit
error rate), and the channel loss at the configured distance. A positive key rate means
that secure key generation is feasible at this distance.

You can also query the key rate at arbitrary distances and find the maximum secure
distance for this configuration.

```python
# Key rate at specific distances
print(f"Key rate at 10 km: {result.key_rate_at(10):.4e} bits/s")
print(f"Key rate at 30 km: {result.key_rate_at(30):.4e} bits/s")
```

```python
print(f"Maximum secure distance: {result.max_distance_km():.1f} km")
```
