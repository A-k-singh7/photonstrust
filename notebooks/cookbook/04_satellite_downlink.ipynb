# Satellite QKD — LEO Downlink Budget

Satellite-based QKD extends secure key distribution to intercontinental distances by
avoiding the exponential fibre loss. China's **Micius** satellite (2017) demonstrated
BB84 downlinks over 1,200 km with a key rate of several kbits/s, proving the concept
for global quantum-secured communication.

In this notebook we use `plan_satellite` to compute the link budget for a Low Earth
Orbit (LEO) downlink and explore how orbital altitude affects performance.

```python
from photonstrust.easy import plan_satellite
```

```python
plan = plan_satellite(orbit_altitude_km=500, protocol="bb84")
print(plan.summary())
```

## Understanding Satellite Link Budgets

The link budget accounts for **geometric spreading loss** (beam divergence over the
slant range), **atmospheric absorption and turbulence**, telescope coupling efficiency,
and detector performance. Higher orbits increase the slant range but extend the
visibility window per pass.

```python
# Compare two orbital altitudes
plan_600 = plan_satellite(orbit_altitude_km=600, protocol="bb84")
print("=== 600 km orbit ===")
print(plan_600.summary())
```
