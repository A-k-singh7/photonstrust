# Protocol Shootout — Comparing All 9 QKD Protocols

Different QKD protocols offer different trade-offs between key rate, maximum distance,
and implementation complexity. PhotonsTrust supports nine protocols out of the box.
This notebook pits several of them against each other so you can see which protocol
wins at each distance.

```python
from photonstrust.easy import compare_protocols
```

```python
comp = compare_protocols(
    protocols=["bb84", "bbm92", "cv_qkd"],
    distances={"start": 0, "stop": 100, "step": 5},
    include_uncertainty=False,
)
```

```python
print(comp.summary())
```

```python
print(f"Best at 20 km: {comp.winner_at(20)}")
print(f"Best at 80 km: {comp.winner_at(80)}")
```

## When to Use Which Protocol

| Protocol | Strengths | Best range |
|----------|-----------|------------|
| **BB84** | Mature, well-understood, commercially deployed | Metro (< 100 km) |
| **BBM92** | Entanglement-based, device-independent potential | Metro |
| **CV-QKD** | Uses standard telecom components (homodyne/heterodyne) | Short-range (< 50 km) |
| **TF-QKD** | Beats the PLOB bound for single-repeater-less links | Long-haul (100–300 km) |
| **SNS-TF-QKD** | Practical twin-field variant with simpler phase tracking | Long-haul |

## Try It Yourself

- Add `"tf_qkd"` and `"sns_tf_qkd"` to the protocol list and extend the distance

- Use `comp.winner_at(d)` in a loop to find the crossover distance where one protocol
  overtakes another.
