# Building a City-Scale QKD Network

Real-world QKD deployments connect multiple sites through a mesh of fibre links.
In a **trusted-node architecture**, intermediate nodes are physically secured and
relay keys between endpoints. This notebook designs a small metro network with four
nodes and analyses its end-to-end key distribution capacity.

```python
from photonstrust.easy import plan_network
```

```python
result = plan_network(
    nodes=["City_Hall", "Bank_HQ", "Data_Center", "Hospital"],
    links=[
        {"a": "City_Hall", "b": "Bank_HQ", "distance_km": 15},
        {"a": "Bank_HQ", "b": "Data_Center", "distance_km": 10},
        {"a": "Data_Center", "b": "Hospital", "distance_km": 20},
        {"a": "City_Hall", "b": "Data_Center", "distance_km": 25},
    ],
    protocol="bb84",
)
```

```python
print(result.summary())
```

## Understanding Network Metrics

The summary reports per-link key rates, the network's **min-cut capacity** (the
bottleneck), and whether every node pair can establish a key via the available paths.
Redundant links improve resilience: if one fibre is cut, alternative routes remain.
